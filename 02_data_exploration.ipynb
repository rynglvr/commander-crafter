```python
# MTG Commander Recommendation System - Step 2: Data Loading and Exploration

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict
import ast
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("üÉè MTG Commander Recommendation System - Step 2")
print("üìä Data Loading and Exploration")
print("=" * 60)

# =============================================================================
# 1. LOAD THE DATA
# =============================================================================

print("\n1Ô∏è‚É£ LOADING DATA...")

# Load the recommendation data
try:
    recommendations_df = pd.read_csv('data/raw/creature_recommendations_clean.csv')
    print(f"‚úÖ Loaded recommendations: {recommendations_df.shape[0]:,} rows, {recommendations_df.shape[1]} columns")
except FileNotFoundError:
    print("‚ùå Could not find creature_recommendations_clean.csv in data/raw/")
    print("Please make sure the file is in the correct location")

# Load the creature data
try:
    creatures_df = pd.read_csv('data/raw/all_creatures_clean.csv')
    print(f"‚úÖ Loaded creatures: {creatures_df.shape[0]:,} rows, {creatures_df.shape[1]} columns")
except FileNotFoundError:
    print("‚ùå Could not find all_creatures_clean.csv in data/raw/")
    print("Please make sure the file is in the correct location")

# =============================================================================
# 2. BASIC DATA EXPLORATION
# =============================================================================

print("\n2Ô∏è‚É£ BASIC DATA EXPLORATION...")

print("\nüìã RECOMMENDATIONS DATA STRUCTURE:")
print(recommendations_df.head())
print(f"\nColumns: {list(recommendations_df.columns)}")
print(f"\nData types:\n{recommendations_df.dtypes}")
print(f"\nMissing values:\n{recommendations_df.isnull().sum()}")

print("\nüìã CREATURES DATA STRUCTURE:")
print(creatures_df.head())
print(f"\nColumns: {list(creatures_df.columns)}")
print(f"\nData types:\n{creatures_df.dtypes}")
print(f"\nMissing values:\n{creatures_df.isnull().sum()}")

# =============================================================================
# 3. ANALYZE RECOMMENDATIONS DATA
# =============================================================================

print("\n3Ô∏è‚É£ ANALYZING RECOMMENDATIONS...")

# Count recommendations per commander
recs_per_commander = recommendations_df.groupby('commander_name').size().sort_values(ascending=False)
print(f"\nüìä Recommendations per Commander (Top 10):")
print(recs_per_commander.head(10))

# Synergy score distribution
print(f"\nüìä Synergy Score Distribution:")
print(recommendations_df['synergy_percentage'].describe())

# Plot synergy distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(recommendations_df['synergy_percentage'], bins=20, alpha=0.7, color='steelblue')
plt.title('Distribution of Synergy Scores')
plt.xlabel('Synergy Percentage')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.boxplot(recommendations_df['synergy_percentage'])
plt.title('Synergy Score Box Plot')
plt.ylabel('Synergy Percentage')

plt.tight_layout()
plt.show()

# =============================================================================
# 4. JOIN DATA AND ANALYZE COLOR IDENTITY
# =============================================================================

print("\n4Ô∏è‚É£ JOINING DATA AND ANALYZING COLOR IDENTITY...")

def parse_color_identity(color_str):
    """Parse color identity string into list - Updated for new format"""
    try:
        if pd.isna(color_str) or color_str == '[]':
            return []
        # Handle string representation of list
        if isinstance(color_str, str):
            # Remove brackets and split by comma
            color_str = color_str.strip('[]')
            if not color_str:
                return []
            # Split by comma and clean up spaces
            return [c.strip() for c in color_str.split(',')]
        return color_str
    except:
        return []

# Join recommendations with commander data to get commander color identity
print("üîó Joining recommendations with commander data...")
# Include usd_price in the columns we're selecting
recommendations_with_commander = recommendations_df.merge(
    creatures_df[['name', 'color_identity', 'usd_price']], 
    left_on='commander_name', 
    right_on='name', 
    how='left',
    suffixes=('', '_commander')
)

# Join with recommended creature data to get recommended creature info
print("üîó Joining with recommended creature data...")
recommendations_full = recommendations_with_commander.merge(
    creatures_df[['name', 'color_identity', 'oracle_text', 'power', 'toughness', 'secondary_type', 'keywords', 'usd_price']], 
    left_on='recommended_creature_name', 
    right_on='name', 
    how='left',
    suffixes=('_commander', '_recommended')
)

print(f"‚úÖ Successfully joined data: {recommendations_full.shape[0]:,} rows")

# Check join success
commander_join_success = recommendations_full['color_identity_commander'].notna().sum()
recommended_join_success = recommendations_full['color_identity_recommended'].notna().sum()

print(f"üìä Join Success Rates:")
print(f"  ‚Ä¢ Commander data joined: {commander_join_success:,}/{len(recommendations_df):,} ({commander_join_success/len(recommendations_df)*100:.1f}%)")
print(f"  ‚Ä¢ Recommended creature data joined: {recommended_join_success:,}/{len(recommendations_df):,} ({recommended_join_success/len(recommendations_df)*100:.1f}%)")

# Parse color identities
print("üé® Parsing color identities...")
recommendations_full['commander_colors_parsed'] = recommendations_full['color_identity_commander'].apply(parse_color_identity)
recommendations_full['recommended_colors_parsed'] = recommendations_full['color_identity_recommended'].apply(parse_color_identity)

# Analyze commander color combinations
color_combinations = recommendations_full['commander_colors_parsed'].apply(lambda x: ''.join(sorted(x)) if x else 'Colorless')
color_counts = color_combinations.value_counts()

print(f"üìä Commander Color Combinations:")
print(color_counts.head(10))

# Plot color distribution
plt.figure(figsize=(12, 6))
color_counts.head(15).plot(kind='bar')
plt.title('Most Common Commander Color Combinations')
plt.xlabel('Color Identity')
plt.ylabel('Number of Recommendations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Validate color identity rules
print("\nüéØ Validating Color Identity Rules...")
def is_color_subset(recommended_colors, commander_colors):
    """Check if recommended creature colors are subset of commander colors"""
    if not recommended_colors:  # Colorless is always valid
        return True
    if not commander_colors:  # Colorless commander can only have colorless creatures
        return False
    return set(recommended_colors).issubset(set(commander_colors))

# Check color identity compliance
recommendations_full['color_identity_valid'] = recommendations_full.apply(
    lambda row: is_color_subset(row['recommended_colors_parsed'], row['commander_colors_parsed']), 
    axis=1
)

valid_count = recommendations_full['color_identity_valid'].sum()
total_count = len(recommendations_full)
print(f"Color identity compliance: {valid_count:,}/{total_count:,} ({valid_count/total_count*100:.1f}%)")

if valid_count < total_count:
    invalid_examples = recommendations_full[~recommendations_full['color_identity_valid']].head(5)
    print("\nSample invalid color combinations:")
    for _, row in invalid_examples.iterrows():
        print(f"  ‚Ä¢ {row['commander_name']} {row['commander_colors_parsed']} ‚Üí {row['recommended_creature_name']} {row['recommended_colors_parsed']}")

# Store the full joined dataframe for later use
print("\nüíæ Storing joined dataframe for later steps...")
recommendations_full.to_csv('data/processed/recommendations_with_features.csv', index=False)
print("‚úÖ Saved to data/processed/recommendations_with_features.csv")

# =============================================================================
# 5. ANALYZE CREATURE FEATURES
# =============================================================================

print("\n5Ô∏è‚É£ ANALYZING CREATURE FEATURES...")

# Check if we can join the data
commander_names = set(recommendations_df['commander_name'].unique())
creature_names = set(creatures_df['name'].unique())

print(f"üìä Data Join Analysis:")
print(f"Unique commanders: {len(commander_names):,}")
print(f"Unique creatures: {len(creature_names):,}")
print(f"Commanders found in creatures data: {len(commander_names & creature_names):,}")

# Find missing commanders
missing_commanders = commander_names - creature_names
if missing_commanders:
    print(f"‚ö†Ô∏è Missing commanders ({len(missing_commanders)}): {list(missing_commanders)[:10]}")

# Analyze recommended creatures
recommended_names = set(recommendations_df['recommended_creature_name'].unique())
print(f"Unique recommended creatures: {len(recommended_names):,}")
print(f"Recommended creatures found in data: {len(recommended_names & creature_names):,}")

# Find missing recommended creatures
missing_recommended = recommended_names - creature_names
if missing_recommended:
    print(f"‚ö†Ô∏è Missing recommended creatures ({len(missing_recommended)}): {list(missing_recommended)[:10]}")

# =============================================================================
# 6. ANALYZE KEYWORDS AND SECONDARY TYPES FROM JOINED DATA
# =============================================================================

print("\n6Ô∏è‚É£ ANALYZING KEYWORDS AND SECONDARY TYPES...")

# Check keywords from recommended creatures
if 'keywords' in recommendations_full.columns:
    print("‚úÖ Found 'keywords' column in joined data")
    
    # Parse keywords function - Updated for new format
    def parse_keywords(keywords_str):
        """Parse keywords string into list - Updated for new format"""
        try:
            if pd.isna(keywords_str) or keywords_str == '[]':
                return []
            if isinstance(keywords_str, str):
                # Remove brackets and split by comma
                keywords_str = keywords_str.strip('[]')
                if not keywords_str:
                    return []
                # Split by comma and clean up spaces
                return [k.strip() for k in keywords_str.split(',')]
            return keywords_str
        except:
            return []
    
    # Parse keywords
    recommendations_full['keywords_parsed'] = recommendations_full['keywords'].apply(parse_keywords)
    
    # Sample keywords
    sample_keywords = recommendations_full['keywords'].dropna().head(10)
    print("Sample keywords (raw format):")
    for i, kw in enumerate(sample_keywords):
        print(f"  {i+1}. {kw}")
        
    # Count all keywords across all recommendations
    all_keywords = []
    for keywords_list in recommendations_full['keywords_parsed']:
        if keywords_list:
            all_keywords.extend(keywords_list)
    
    keyword_counts = Counter(all_keywords)
    print(f"\nTop 15 most common keywords:")
    for keyword, count in keyword_counts.most_common(15):
        print(f"  {keyword}: {count:,}")
        
else:
    print("‚ùå 'keywords' column not found in joined data")

# Check secondary types
if 'secondary_type' in recommendations_full.columns:
    print("\n‚úÖ Found 'secondary_type' column in joined data")
    
    # Parse secondary types - Updated for new format
    def parse_secondary_types(secondary_type_str):
        """Parse secondary types string into list - Updated for new format"""
        try:
            if pd.isna(secondary_type_str) or secondary_type_str == '[]':
                return []
            if isinstance(secondary_type_str, str):
                # Remove brackets and split by comma
                secondary_type_str = secondary_type_str.strip('[]')
                if not secondary_type_str:
                    return []
                # Split by comma and clean up spaces
                return [t.strip() for t in secondary_type_str.split(',')]
            return secondary_type_str
        except:
            return []
    
    # Parse secondary types
    recommendations_full['secondary_types_parsed'] = recommendations_full['secondary_type'].apply(parse_secondary_types)
    
    # Sample secondary types (original format)
    sample_secondary_types = recommendations_full['secondary_type'].dropna().head(10)
    print("Sample secondary types (original format):")
    for i, st in enumerate(sample_secondary_types):
        print(f"  {i+1}. '{st}'")
    
    print("\nSample secondary types (parsed format):")
    for i, st in enumerate(recommendations_full['secondary_types_parsed'].head(10)):
        print(f"  {i+1}. {st}")
    
    # Count all individual secondary types across all recommendations
    all_secondary_types = []
    for types_list in recommendations_full['secondary_types_parsed']:
        if types_list:
            all_secondary_types.extend(types_list)
    
    secondary_type_counts = Counter(all_secondary_types)
    print(f"\nTop 20 most common individual secondary types:")
    for sec_type, count in secondary_type_counts.most_common(20):
        print(f"  {sec_type}: {count:,}")
    
    # Also show the full secondary_type combinations
    combination_counts = recommendations_full['secondary_type'].value_counts().head(15)
    print(f"\nTop 15 secondary type combinations:")
    print(combination_counts)
    
    # Plot individual secondary type distribution
    plt.figure(figsize=(15, 10))
    
    plt.subplot(2, 1, 1)
    top_individual = pd.Series(dict(secondary_type_counts.most_common(20)))
    top_individual.plot(kind='bar')
    plt.title('Most Common Individual Secondary Types')
    plt.xlabel('Secondary Type')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    
    plt.subplot(2, 1, 2)
    combination_counts.head(15).plot(kind='bar')
    plt.title('Most Common Secondary Type Combinations')
    plt.xlabel('Secondary Type Combination')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
else:
    print("‚ùå 'secondary_type' column not found in joined data")

# =============================================================================
# 7. ORACLE TEXT ANALYSIS FROM JOINED DATA
# =============================================================================

print("\n7Ô∏è‚É£ ANALYZING ORACLE TEXT...")

if 'oracle_text' in recommendations_full.columns:
    print("‚úÖ Found 'oracle_text' column in joined data")
    
    # Basic oracle text stats
    oracle_stats = recommendations_full['oracle_text'].str.len().describe()
    print("Oracle text length statistics:")
    print(oracle_stats)
    
    # Sample oracle texts from recommended creatures
    print("\nSample oracle texts from recommended creatures:")
    sample_oracle = recommendations_full['oracle_text'].dropna().head(5)
    for i, text in enumerate(sample_oracle):
        print(f"  {i+1}. {text[:150]}...")
        
    # Plot oracle text length distribution
    plt.figure(figsize=(10, 6))
    recommendations_full['oracle_text'].str.len().hist(bins=30, alpha=0.7)
    plt.title('Distribution of Oracle Text Lengths')
    plt.xlabel('Oracle Text Length (characters)')
    plt.ylabel('Frequency')
    plt.show()
        
else:
    print("‚ùå 'oracle_text' column not found in joined data")

# =============================================================================
# 8. POWER/TOUGHNESS ANALYSIS FROM JOINED DATA
# =============================================================================

print("\n8Ô∏è‚É£ ANALYZING POWER/TOUGHNESS...")

if 'power' in recommendations_full.columns and 'toughness' in recommendations_full.columns:
    print("‚úÖ Found power and toughness columns in joined data")
    
    # Convert to numeric, handling special values like '*'
    def clean_numeric(val):
        try:
            if pd.isna(val) or val == '*':
                return np.nan
            return float(val)
        except:
            return np.nan
    
    recommendations_full['power_numeric'] = recommendations_full['power'].apply(clean_numeric)
    recommendations_full['toughness_numeric'] = recommendations_full['toughness'].apply(clean_numeric)
    
    # Analyze power/toughness distribution
    print(f"\nPower statistics:")
    print(recommendations_full['power_numeric'].describe())
    print(f"\nToughness statistics:")
    print(recommendations_full['toughness_numeric'].describe())
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    recommendations_full['power_numeric'].hist(bins=20, alpha=0.7)
    plt.title('Power Distribution (Recommended Creatures)')
    plt.xlabel('Power')
    plt.ylabel('Frequency')
    
    plt.subplot(1, 3, 2)
    recommendations_full['toughness_numeric'].hist(bins=20, alpha=0.7)
    plt.title('Toughness Distribution (Recommended Creatures)')
    plt.xlabel('Toughness')
    plt.ylabel('Frequency')
    
    plt.subplot(1, 3, 3)
    plt.scatter(recommendations_full['power_numeric'], recommendations_full['toughness_numeric'], alpha=0.5)
    plt.title('Power vs Toughness')
    plt.xlabel('Power')
    plt.ylabel('Toughness')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("‚ùå Power/toughness columns not found in joined data")

# =============================================================================
# 9. PRICE ANALYSIS FROM JOINED DATA (NEW)
# =============================================================================

print("\n9Ô∏è‚É£ ANALYZING CARD PRICES...")

if 'usd_price_commander' in recommendations_full.columns and 'usd_price_recommended' in recommendations_full.columns:
    print("‚úÖ Found USD price columns in joined data")
    
    # Analyze commander prices
    print(f"\nCommander Price Statistics:")
    print(recommendations_full['usd_price_commander'].describe())
    
    # Analyze recommended creature prices
    print(f"\nRecommended Creature Price Statistics:")
    print(recommendations_full['usd_price_recommended'].describe())
    
    # Create price visualizations
    plt.figure(figsize=(15, 10))
    
    # Distribution of commander prices
    plt.subplot(2, 2, 1)
    recommendations_full['usd_price_commander'].dropna().hist(bins=30, alpha=0.7)
    plt.title('Commander Price Distribution')
    plt.xlabel('USD Price')
    plt.ylabel('Frequency')
    plt.xlim(0, recommendations_full['usd_price_commander'].quantile(0.95))  # Limit to 95th percentile for better visualization
    
    # Distribution of recommended creature prices
    plt.subplot(2, 2, 2)
    recommendations_full['usd_price_recommended'].dropna().hist(bins=30, alpha=0.7)
    plt.title('Recommended Creature Price Distribution')
    plt.xlabel('USD Price')
    plt.ylabel('Frequency')
    plt.xlim(0, recommendations_full['usd_price_recommended'].quantile(0.95))  # Limit to 95th percentile
    
    # Box plots for price comparison
    plt.subplot(2, 2, 3)
    price_data = [
        recommendations_full['usd_price_commander'].dropna(),
        recommendations_full['usd_price_recommended'].dropna()
    ]
    plt.boxplot(price_data, labels=['Commanders', 'Recommended Creatures'])
    plt.title('Price Comparison Box Plot')
    plt.ylabel('USD Price')
    plt.ylim(0, 50)  # Limit y-axis for better visualization
    
    # Average recommended creature price by commander price range
    plt.subplot(2, 2, 4)
    # Create price bins for commanders
    commander_price_bins = pd.cut(recommendations_full['usd_price_commander'], 
                                  bins=[0, 5, 10, 20, 50, 100, float('inf')],
                                  labels=['$0-5', '$5-10', '$10-20', '$20-50', '$50-100', '$100+'])
    avg_rec_price_by_commander_bin = recommendations_full.groupby(commander_price_bins)['usd_price_recommended'].mean()
    avg_rec_price_by_commander_bin.plot(kind='bar')
    plt.title('Average Recommended Creature Price by Commander Price Range')
    plt.xlabel('Commander Price Range')
    plt.ylabel('Average Recommended Creature Price (USD)')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Top 10 most expensive commanders with recommendations
    expensive_commanders = recommendations_full.groupby('commander_name')['usd_price_commander'].first().sort_values(ascending=False).head(10)
    print("\nTop 10 Most Expensive Commanders:")
    for commander, price in expensive_commanders.items():
        print(f"  {commander}: ${price:.2f}")
    
    # Top 10 most expensive recommended creatures
    expensive_recommendations = recommendations_full.groupby('recommended_creature_name')['usd_price_recommended'].first().sort_values(ascending=False).head(10)
    print("\nTop 10 Most Expensive Recommended Creatures:")
    for creature, price in expensive_recommendations.items():
        print(f"  {creature}: ${price:.2f}")
    
    # Analyze price correlation with synergy
    print("\nüìä Price vs Synergy Analysis:")
    price_synergy_corr = recommendations_full[['usd_price_recommended', 'synergy_percentage']].corr()
    print("Correlation between recommended creature price and synergy score:")
    print(price_synergy_corr)
    
    # Scatter plot of price vs synergy
    plt.figure(figsize=(10, 6))
    plt.scatter(recommendations_full['synergy_percentage'], 
                recommendations_full['usd_price_recommended'], 
                alpha=0.3)
    plt.xlabel('Synergy Percentage')
    plt.ylabel('Recommended Creature Price (USD)')
    plt.title('Recommended Creature Price vs Synergy Score')
    plt.ylim(0, recommendations_full['usd_price_recommended'].quantile(0.95))
    plt.show()
    
else:
    print("‚ùå USD price columns not found in joined data")
    print("Available columns:")
    print([col for col in recommendations_full.columns if 'price' in col.lower()])

# =============================================================================
# 10. SUMMARY AND NEXT STEPS
# =============================================================================

print("\nüîü SUMMARY AND NEXT STEPS...")
print("=" * 60)
print("‚úÖ COMPLETED:")
print("  ‚Ä¢ Data successfully loaded with new format")
print("  ‚Ä¢ Basic exploration completed")
print("  ‚Ä¢ Column mapping identified")
print("  ‚Ä¢ Data quality assessed")
print("  ‚Ä¢ Price analysis added")

print("\nüîç KEY FINDINGS:")
print(f"  ‚Ä¢ {len(commander_names):,} unique commanders")
print(f"  ‚Ä¢ {len(recommended_names):,} unique recommended creatures") 
print(f"  ‚Ä¢ {recommendations_df.shape[0]:,} total recommendations")
print(f"  ‚Ä¢ Synergy scores range: {recommendations_df['synergy_percentage'].min():.2f} - {recommendations_df['synergy_percentage'].max():.2f}")

if 'usd_price_recommended' in recommendations_full.columns:
    avg_price = recommendations_full['usd_price_recommended'].mean()
    median_price = recommendations_full['usd_price_recommended'].median()
    print(f"  ‚Ä¢ Average recommended creature price: ${avg_price:.2f}")
    print(f"  ‚Ä¢ Median recommended creature price: ${median_price:.2f}")

print("\nüöÄ READY FOR STEP 3:")
print("  ‚Ä¢ Data preprocessing and cleaning")
print("  ‚Ä¢ Feature engineering")
print("  ‚Ä¢ Text processing for oracle_text")
print("  ‚Ä¢ Color identity validation")
print("  ‚Ä¢ Price-based feature engineering")

print("\nüíæ Don't forget to save this notebook as '02_data_exploration.ipynb'")
print("=" * 60)
```
